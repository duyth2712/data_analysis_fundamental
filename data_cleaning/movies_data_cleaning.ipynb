```python
#Find out the percentage of missing values in each column
percent_missing = df.isnull().sum() * 100 / len(df)
missing_value_df = pd.DataFrame({'column_name': df.columns,
                                 'percent_missing': percent_missing})
print(missing_value_df)
```

```python
#fill and clean YEAR column
df['YEAR'] = df['YEAR'].str.extract(r'([0-9]{4}–.*|[0-9]{4})')
df['YEAR'] = df['YEAR'].str.strip().replace(")","")

def extract_year(year):
    if year[-3:] == '– )':
        return year.replace('– )',"-")
    else:
        return year.replace(')',"")

df['YEAR'] = df['YEAR'].fillna('Unknown')
df['YEAR'] = df['YEAR'].apply(lambda y: extract_year(y))

df['YEAR'].head(10)
```

```python
#clean GENRE columns
df["GENRE"]=df["GENRE"].str.replace("\n","")
df['GENRE'] = df['GENRE'].str.strip()
df['GENRE'].fillna('unknown')
df["GENRE"].head(10)
```

```python
#clean ONE-LINE columns
df["ONE-LINE"]=df["ONE-LINE"].str.replace("\n","")
df['ONE-LINE'] = df['ONE-LINE'].str.strip()
df["ONE-LINE"].head(10)
```

```python
# Clean and Split STARS column into DIRECTOR column and STAR column
df['STARS']=df['STARS'].str.replace("\n","")
df['STARS'] = df['STARS'].str.strip()

def extract_director(direc):
    if 'Director' in direc or 'Directors' in direc:
        director = direc.strip().split("|")[0] # The Second Half is the stars
        return director.split(":")[1] # Return the Director name
    else:
        return ''

def extract_stars(stars):
    if 'Star' not in stars or 'Stars' not in stars:
        return ''
    else:
        return stars.split(":")[-1] # last value in this list will be the stars
 
df['DIRECTOR'] = df['STARS'].apply(lambda d: extract_director(d))
df['STAR'] = df['STARS'].apply(lambda s: extract_stars(s))

df[['DIRECTOR','STAR']].head(10)
```

```python
#fill RATING, RunTime columns with mean value
df['RATING'].fillna(int(df['RATING'].mean()),inplace=True)
df['RunTime'].fillna(int(df['RunTime'].mean()),inplace=True)

df.head(10)
```

```python
df['VOTES'] = df['VOTES'].str.replace(",","")
df['VOTES'] 
df['VOTES'] = df['VOTES'].fillna(0)
df['VOTES'] = df['VOTES'].astype(int)
df.head(10)
```
